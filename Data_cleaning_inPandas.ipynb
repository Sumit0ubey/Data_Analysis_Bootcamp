```python
import pandas as pd
```

```python
df = pd.read_excel(r"S:\VS code\python\Data _Analytics\Data_Analysis_Bootcamp\Customer Call List.xlsx")
```

```python
df
```

```python
df1 = df.drop_duplicates()
df1
```

```python
df1 = df1.drop(columns = "Not_Useful_Column")
df1
```

```python
df1['Last_Name'] = df1['Last_Name'].str.strip('123./')
df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].str.replace('nan--','')
df1["Phone_Number"] = df1["Phone_Number"].str.replace('Na--','')
df1["Phone_Number"] = df1["Phone_Number"].str.replace('N/a','')
df1['Phone_Number'] = df1['Phone_Number'].fillna('')

df1["Phone_Number"] = df1["Phone_Number"].str.replace('|','-')
df1["Phone_Number"] = df1["Phone_Number"].str.replace('/','-')
df1
```

```python
df1["Do_Not_Contact"] = df1["Do_Not_Contact"].str.replace('Yes','Y')
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('No', 'N')
df1
```

```python
for x in df1.index:
    if (df1.loc[x, 'Do_Not_Contact'] == 'Y') or (df1.loc[x, 'Phone_Number'] == ''):
        df1.drop(x, inplace=True)

df1
```

```python
df1 = df1.reset_index(drop=True)
df1
```
