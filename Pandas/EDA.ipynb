```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
plt.bar(df['Continent'], df['2015 Population'])
plt.title('Continent v/s 2015 Population ')
plt.ylabel('2015 Population')
plt.show()
```

```python
sns.set(style="whitegrid")

# 1. Bar Plot: Top 10 most populous countries in 2022
top_10_populous = df.nlargest(10, '2022 Population')
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='2022 Population', data=top_10_populous, palette='viridis')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Pie Chart: Population distribution by continent in 2022
continent_population = df.groupby('Continent')['2022 Population'].sum()
plt.figure(figsize=(8, 8))
plt.pie(continent_population, labels=continent_population.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Population Distribution by Continent in 2022')
plt.show()
```

```python
# 3. Line Plot: Population growth over the years for a selected country (e.g., India)
india_population = df[df['Country'] == 'India'].iloc[0]
years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', 
         '2010 Population', '2015 Population', '2020 Population', '2022 Population']
plt.figure(figsize=(10, 6))
plt.plot(years, india_population[years], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.title('Population Growth Over the Years for India')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 4. Scatter Plot: Population density vs. population size
plt.figure(figsize=(10, 6))
sns.scatterplot(x='2022 Population', y='Density (per km²)', hue='Continent', size='2022 Population',
                data=df, palette='muted', sizes=(20, 200))
plt.title('Population Density vs. Population Size in 2022')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Density (per km²) (log scale)')
plt.show()
```
